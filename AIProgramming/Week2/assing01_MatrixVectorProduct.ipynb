## Example 01-1 (Matrix-Vector Multiplication)
Write code for the follwoing problem.
\begin{equation}
\begin{bmatrix}
1 & 4 & 2 & 0 \\
9 & 5 & 0 & 0 \\
4 & 0 & 2 & 4 \\
6 & 1 & 8 & 3
\end{bmatrix}
\begin{bmatrix}
1\\2\\3\\4
\end{bmatrix}=?
\end{equation}

```python
A = np.matrix('1, 4, 2, 0; 9, 5, 0, 0; 4, 0, 2, 4; 6, 1, 8, 3')
x = np.matrix('1;2;3;4')
b = np.zeros((4,1))
n = 4

for i in range(0, n):
    val = 0.0
    for j in range(0,n):
        # To-do 1
        val += A[i, j] * x[j]
    b[i] = val
print(b)
```

```python
A = np.array([[1, 4, 2, 0], [9, 5, 0, 0], [4, 0, 2, 4], [6, 1, 8, 3]])
x = np.array([1,2,3,4])
b = np.array([0,0,0,0])
n = 4
for i in range(0, n):
    val = 0.0
    for j in range(0,n):
        # To-do 2
        val += A[i, j] * x[j]
    b[i] = val
print(b)
```

## Example 02-1 (Matrix-Vector Multiplication)
Write code for the follwoing problem.
\begin{equation}
\begin{bmatrix}
4 & 5 & 2 & 1 \\
2 & 3 & 8 & 0 \\
1 & 0 & 7 & 2
\end{bmatrix}
\begin{bmatrix}
1\\2\\3\\4
\end{bmatrix}=?
\end{equation}

```python
A = np.matrix('4, 5, 2, 1; 2, 3, 8, 0; 1, 0, 7, 2')
x = np.matrix('1;2;3;4')
m = 3
n = 4
b = np.zeros((m,1))
# TODO 5
for i in range(0, m):
    val = 0
    for j in range(0,n):
      val += A[i,j] * x[j]
    b[i] = val
print(b)
```
