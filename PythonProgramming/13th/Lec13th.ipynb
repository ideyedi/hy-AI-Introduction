```python
import time
import traceback
import pandas as pd
from selenium import webdriver


d = webdriver.Chrome('/Users/taehwalee/Downloads/chromedriver')

try:
    url = 'https://twitter.com/search?q=python&src=typed_query'
    d.get(url)
    
    time.sleep(3)
    contents = []
    
    data = []
    for _ in range(5):
        elems = d.find_elements_by_css_selector('article[role=article]')
        for elem in elems:
            try:
                content_tag = elem.find_element_by_css_selector('div[lang]')
            except:
                continue

            if content_tag.text in contents:
                continue
                
            author_tag = elem.find_element_by_css_selector('div[dir=ltr] > span')
            time_tag = elem.find_element_by_tag_name('time')
            contents.append(content_tag.text)
            
            data.append(
                (
                    author_tag.text.replace(',', '.').replace('\n', ' '),
                    content_tag.text.replace(',', '.').replace('\n', ' '),
                    time_tag.get_attribute('datetime')
                )
            )
            
        d.execute_script('window.scrollTo(0, document.body.scrollHeight);')
        time.sleep(4)
        
    # 저장하는 코드
    with open('twitter.csv', 'w') as f:
        f.write('username,contents,datetime\n')
        for row in data:
            row_str = ','.join(row)
            f.write(row_str + '\n')
except:
    traceback.print_exc()
finally:
    time.sleep(3)
    d.quit()
    
df = pd.read_csv('twitter.csv')
df.datetime = pd.to_datetime(df.datetime)
df['year'] = df.datetime.dt.year
df['month'] = df.datetime.dt.month
df
```
