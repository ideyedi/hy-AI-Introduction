```python
'''
If there is an import error, please run the command below
> pip install pandas numpy
'''
import pandas as pd
import numpy as np
import datetime

file_path = './large_bike.csv'

# Read csv file
df_tot = pd.read_csv(file_path)

# make error-checklist
checklist = df_tot.location.unique()

location = input('input:')
if location not in checklist:
    location = None
    print('Error input value\n input must be in {}'.format(checklist))
```

```python
def MakePandasReport(df):
    # remake dateframe
    df = df.loc[:, ['datetime', 'casual', 'registered', 'location']]
    # transform string to datetime object which have a date
    df['datetime'] = pd.to_datetime(df['datetime']).dt.date
    # reorganize based on location
    df = df[df.location == location]
    # sort by datetime and reset index
    df = df.sort_values('datetime').reset_index(drop=True)
    # get total count by datetime
    df = df.groupby('datetime').agg('sum')

    return df

if 'df_tot' in globals():
    df = MakePandasReport(df_tot)
else:
    print('plz, Run first sequence')
```

```python
# function that parse csv file
def csvParser(file):
    data_list = []
    with open(file, 'r', encoding='UTF-8') as f:
        while True:
            line = f.readline().strip()
            if not line:
                break
            
            columns = line.split(',')
            data_list.append(columns)
    
    return data_list

def MakeDicReport(dic):
    # make large_bike matrix
    lbs = csvParser(file_path)

    date_list = []
    lo_matrix = []

    for lb in lbs:
        if lb[12] == location:
            tmp = []
            tmp.append(lb[0][0:10])
            tmp.append(lb[9])
            tmp.append(lb[10])
            tmp.append(lb[12])
            date_list.append(lb[0][0:10])
            lo_matrix.append(tmp)

    lo_matrix.sort(key=lambda x:x[0])

    # remove duplicate date, sort list
    unique_date = list(set(date_list))
    unique_date.sort()

    # make dictionary
    for i in lo_matrix:
        try:
            dic[i[0]+'-casual'] += int(i[1])
            dic[i[0]+'-registered'] += int(i[2])
        
        except:
            dic[i[0]+'-casual'] = int(i[1])
            dic[i[0]+'-registered'] = int(i[2])
    #call by assignment..
    return unique_date

dic = {}
unique_date = []

if 'file_path' in globals():
    unique_date = MakeDicReport(dic)
    #print(dic, unique_date)
else:
    print('plz, run first sequence')
```

```python
import os

def PrintDicReport():
    for date in unique_date:
        print('{}: casual:{}, registered:{}'.format(date, dic[date+'-casual'], dic[date+'-registered']))

if 'location' and 'unique_date' not in globals():
    os._exit(0)
    
print('input:{}'.format(location))
PrintDicReport()
print('-'*40)
df
```
