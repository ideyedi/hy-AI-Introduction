```python
import pandas as pd
import numpy as np
df = pd.read_csv('../5th/bike.csv')
df_sub = df.iloc[:6, 9:12]
df_sub
```

```python
# sort_valuse default ordering is increasment
df_sub = df_sub.sort_values('registered')
df_sub
```

```python
# sorting 시 index가 꼬여있는 부분이 중요
df_sub = df_sub.sort_values('registered', ascending=False)
df_sub
```

```python
# 다중 필드 정렬
# multiple feature sorting
# priority 1 casual, 2 registered
df_sub = df_sub.sort_values(['casual', 'registered'])
df_sub
```

```python
# ascending, ordering method도 list 형태로 선택이 가능하다
df_sub = df_sub.sort_values(['casual', 'registered'], ascending=[False, True])
df_sub
```

```python
# index가 다시 재정립되지만, index feature가 생긴다...
# 왜 이따구로 만들었는가 ㅋㅋㅋ
df_reset = df_sub.reset_index()
df_reset
```

```python
# index feature 생성, 삭제 여부를 정할 수 있다.
df_drop = df_sub.reset_index(drop=True)
df_drop
```

```python
df_sub = df.loc[:5, 'casual':"count"]
# Add index feature
df_sub['idx'] = df_sub.index
df_sub
```

```python
# 'idx'를 기준으로 한 행으로 테이블이 구성되는 method 'melt'
# 기준 값이 list로 들어오는 경우도 있을 수 있으며 이때는 해당 리스트의 값을 기준으로 행을 재구성함
df_sub.melt(id_vars='idx')
```

```python
#Normalize 정규화
df_sub = df.loc[:, ['casual', 'registered']]
df_sub['casual2'] = (df_sub["casual"] - min(df_sub["casual"])) / (max(df_sub['casual']) - min(df_sub['casual']))
df_sub
```

```python
#Normalize 표준화
df_sub = df.loc[:, ['casual', 'registered']]
casual = df_sub['casual']
df_sub['casual'] = (casual - casual.mean()) / casual.std()
df_sub.head(10)
```
