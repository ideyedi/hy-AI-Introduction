```python
import time
import traceback
from selenium import webdriver
# key-binding
from selenium.webdriver.common.keys import Keys
```

```python
# options
#op = webdriver.ChromeOptions()
#op.add_argument('--headless')
#op.add_argument('--no-sandbox')
#op.add_argument('--disable-dev-shm-usage')

# chromedriver version 90
d = webdriver.Chrome('chromedriver')

try:
    d.get('https://www.naver.com')
    
    elem = d.find_element_by_id('query')
    elem.send_keys('Hanyang')
    elem.send_keys(Keys.RETURN)
    
    # hanyang pages
    elems = d.find_elements_by_class_name('link_tit')
    
    for elem in elems:
        print(elem.text, elem.get_attribute('href'))
    
except:
    #print(e)
    traceback.print_exc()
finally:
    time.sleep(5)
    d.quit()
```

```python
# Addtional pagenation crawling
# chromedriver version 90
d = webdriver.Chrome('chromedriver')

try:
    d.get('https://cafe.naver.com/joonggonara')
    
    elem = d.find_element_by_id('topLayerQueryInput')
    elem.send_keys('RTX 3080')
    elem.send_keys(Keys.RETURN)
    
    iframe = d.find_element_by_id('cafe_main')
    d.switch_to.frame(iframe)
    
    idx = 2
    while True:
        elems = d.find_elements_by_class_name('article')
        #for elem in elems:
        #    print(elem.text, elem.get_attribute('href'))

        if idx % 10 == 1:
            next_page = d.find_element_by_class_name('pgR')
        else:
            prev_next = d.find_element_by_class_name('prev-next')
            next_page = prev_next.find_element_by_link_text(str(idx))
        
        next_page.click()
        idx += 1
    
except:
    #print(e)
    traceback.print_exc()
finally:
    time.sleep(5)
    d.quit()
```

```python
# Addtional get content
# Addtional pagenation crawling
# chromedriver version 90
d = webdriver.Chrome('chromedriver')
links = []

try:
    d.get('https://cafe.naver.com/joonggonara')
    
    elem = d.find_element_by_id('topLayerQueryInput')
    elem.send_keys('RTX 3080')
    elem.send_keys(Keys.RETURN)
    
    iframe = d.find_element_by_id('cafe_main')
    d.switch_to.frame(iframe)
    
    idx = 2
    while True:
        elems = d.find_elements_by_class_name('article')
        for elem in elems:
            #print(elem.text, elem.get_attribute('href'))
            links.append(elem.get_attribute('href'))

        if idx % 10 == 1:
            next_page = d.find_element_by_class_name('pgR')
        else:
            prev_next = d.find_element_by_class_name('prev-next')
            next_page = prev_next.find_element_by_link_text(str(idx))
        
        next_page.click()
        idx += 1
        
        if idx > 5:
            #print(links)
            break
            
    for link in links:
        d.switch_to.default_content()
        d.get(link)
        time.sleep(2)
        iframe = d.find_element_by_id('cafe_main')
        d.switch_to.frame(iframe)
        
        elem = d.find_element_by_class_name('cost')
        print(elem.text)
        
except:
    #print(e)
    traceback.print_exc()
finally:
    time.sleep(5)
    d.quit()
```
