```python
from selenium import webdriver
import traceback
import time
```

```python
d = webdriver.Chrome('chromedriver')
try : 
    d.get('https://www.youtube.com/channel/UC3SyT4_WLHzN7JmHQwKQZww')
    time.sleep(5)
    
    channel_header = d.find_element_by_id('inner-header-container')
    channel_name = channel_header.find_element_by_id('channel-name').text
    subscriber = channel_header.find_element_by_id('subscriber-count').text.split(' ')[-1]
    print("name {} sub {}".format(channel_name, subscriber))

    tabContents = d.find_elements_by_class_name('tab-content')    
    video_btn = tabContents[1]
    video_btn.click()
    time.sleep(5)
    
    scroll_page_height = d.execute_script('return document.documentElement.scrollHeight')
    while True:
        d.execute_script('window.scrollTo(0, document.documentElement.scrollHeight)')
        time.sleep(2)
        new_scroll_page_height = d.execute_script('return document.documentElement.scrollHeight')
        if scroll_page_height == new_scroll_page_height:
            break
        
        scroll_page_height = new_scroll_page_height
    
    
    video_tag_list = d.find_elements_by_css_selector('ytd-grid-renderer ytd-grid-video-renderer')
    print('contents count:', len(video_tag_list))
    print('='*70)
    
    for video_tag in video_tag_list:
        title_tag = video_tag.find_element_by_id('video-title')
        metadata_tags = video_tag.find_elements_by_css_selector('#metadata span')
        view_count_tag, data_tag = metadata_tags[1:]

        view_count = view_count_tag.text.split(' ')[-1]
        thumbnail_tag = video_tag.find_element_by_id('thumbnail')
        link = thumbnail_tag.get_attribute('href')
        
        print("title : {} [{}, {}]".format(title_tag.text, view_count, data_tag.text))
        print('link : {}'.format(link))
    
    
except:
    traceback.print_exc()

finally:
    time.sleep(2)
    d.close()
    d.quit()
```
