#### **YouTube 첫화면에 보이는 동영상 제목 수집**
----

```python
import traceback
import time
from selenium import webdriver

d = webdriver.Chrome('chromedriver')
try:
    d.get('https://youtube.com')
    time.sleep(5)
    
    videos = d.find_elements_by_css_selector('#contents > .ytd-rich-grid-renderer')
    
    for idx, video in enumerate(videos):
        if idx >= 8:
            break;
        #print(idx+1, video.text)
        title_tag = video.find_element_by_id('video-title-link')
        title = title_tag.get_attribute('title')
        link = title_tag.get_attribute('href')
        print("{} : {} ({})".format(idx + 1, title, link))
        
        channel_tag = video.find_element_by_class_name('yt-formatted-string')
        channel = channel_tag.text
        channel_link = channel_tag.get_attribute('href')
        print("channel : {} ({})".format(channel, channel_link))
        
except:
    time.sleep(2)
    traceback.print_exc()
    d.quit()
```


#### **특정 동영상의 댓글 수집**
----

```python
import re
import datetime

d = webdriver.Chrome('chromedriver')
try:
    d.get('https://www.youtube.com/watch?v=u35cyBQpwNo')
    time.sleep(5)
    
    view_count_tag = d.find_element_by_class_name('view-count')
    view_count = view_count_tag.text
    
    # 조회 수에서 숫자만 빼고 싶다.
    # 언어 설정에 따라 prefix가 달라짐
    # print(view_count)
    pattern = re.compile("\d")
    obj = pattern.findall(view_count)
    print("".join(obj))
    
    # video upload time
    date_tag = d.find_element_by_css_selector('#date > yt-formatted-string')
    date = date_tag.text
    dt = datetime.datetime.strptime(date, '%Y. %m. %d.')
    print(dt)
    
    # Get like! count
    button_tags = d.find_elements_by_css_selector('#menu-container #text')
    like_btn_tag = button_tags[0]
    unlike_btn_tag = button_tags[1]
    
    like_count = like_btn_tag.get_attribute('aria-label')
    unlike_count = unlike_btn_tag.get_attribute('aria-label')
    
    print('like {}, unlink {}'.format(like_count[4:-1], unlike_count[4:-1]))
    
    # Get video infomation as 'contents'
    more_btn = d.find_element_by_class_name('more-button')
    more_btn.click()
    time.sleep(2)
    
    try:
        description_tag = d.find_element_by_id('cc')
    except:
        description_tag = d.find_element_by_css_selector('#content > #description')
    
    description = description_tag.text
    print(description)
    
    # ----
    # 댓글 수집 접근법
    
    # 댓글이 동적페이지로 스크롤링에 따라 생성됨
    # 스크롤을 내리고 댓글을 수집하는 방식으로 접근
    # 여기서는 다 내리지 않고 조금만 내린 후 수집하도록 작성
    #----
    time.sleep(3)
    
    comment_count_tag = d.find_elements_by_css_selector('.count-text > span')[1]
    comment_count = comment_count_tag.text.replace(',', '')
    print(comment_count)
    
    # Scrolling
    #d.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    d.execute_script('window.scrollTo(0, 5000);')
    time.sleep(3)
    
    # Get video comments
    comments = d.find_elements_by_tag_name('ytd-comment-thread-renderer')
    for comment in comments:
        content_tag = comment.find_element_by_id('content-text')
        author_tag = comment.find_element_by_id('author-text')
        print(content_tag.text, author_tag.text)
    
except:
    traceback.print_exc()

finally:
    time.sleep(5)
    #d.quit()
```
