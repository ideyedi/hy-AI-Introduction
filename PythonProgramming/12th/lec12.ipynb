```python
import traceback
import time
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.common.action_chains import ActionChains

options = webdriver.ChromeOptions()
options.add_argument('user-data-dir=/home/eyedi/Algorithm/Hanyang-python/12th/HYUChrome')
# options.add_argument('headless')
# options.add_argument('window-size=1920,1080')

d = webdriver.Chrome('chromedriver', options=options)

try:
    d.get('https://www.instagram.com')

    time.sleep(2)
#     d.save_screenshot('test.png')
#     input()
    
    elem = d.find_element_by_class_name('eyXLr')
    ac = ActionChains(d)
    ac.move_to_element(elem)
    ac.click()
    ac.send_keys('#파이썬')
    ac.pause(3)
    ac.move_by_offset(0, 50)
    ac.click()
    ac.perform()
    
    time.sleep(3)
    
    posts = d.find_elements_by_css_selector('.EZdmt .v1Nh3')
    for post in posts:
        post.click()
        time.sleep(1.5)
        
        contents_tag = d.find_element_by_css_selector('.PpGvg .C4VMK > span')
        contents = contents_tag.text

        tags_text = []
        tags = contents_tag.find_elements_by_tag_name('a')

        for tag in tags:
            contents = contents.replace(tag.text, '')
            tags_text.append(tag.text)
        contents = contents.strip()
        
        comments_text = []
        comments_tag = d.find_elements_by_css_selector('.Mr508 .C4VMK > span')
        for comment_tag in comments_tag:
            comments_text.append(comment_tag.text.strip())
            
        like_tag = d.find_element_by_css_selector('.zV_Nj > span')
        like_count = like_tag.text.strip()
        
        date_tag = d.find_element_by_class_name('_1o9PC')
        date = date_tag.get_attribute('title').strip()
                
        print(contents[:10], tags_text[:2], comments_text[:2], like_count, date)
        print('-'*30)
        
        ac = ActionChains(d)
        ac.send_keys(Keys.ESCAPE)
        ac.perform()
        time.sleep(1)

    
except:
    traceback.print_exc()
#finally:
#    time.sleep(5)
#    d.quit()
```
